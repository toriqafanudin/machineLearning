```python
import numpy as np
import matplotlib.pyplot as plt

x_train = np.array([1.0, 2.0])
y_train = np.array([300.0, 500.0])
```

```python
def compute_cost(x, y, a, b):
    n = x.shape[0]
    cost_sum = 0
    for i in range(n):
        f_ab = a * x[i] + b
        cost = (f_ab - y[i]) ** 2
        cost_sum += cost
    total_cost = (1 / (2 * n)) * cost_sum
    return total_cost
```

```python
x_train = np.array([1.0, 1.7, 2.0, 2.5, 3.0, 3.2])
y_train = np.array([250, 300, 480, 430, 630, 730])
```

```python
# Bisa dijadikan untuk nilai a dan b awal, persamaan garisnya bisa dicari dengan melalui 2 titik
plt.plot([x_train[0], x_train[-1]], [y_train[0], y_train[-1]], c='y')

plt.scatter(x_train, y_train, marker='x', c='r')
plt.show()
```

```python
a = 70
b = 180
total_cost = compute_cost(x_train, y_train, a, b)
print(f'total cost = {total_cost}')
```
